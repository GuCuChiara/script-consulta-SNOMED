#* Agregamos refinamiento con ECL a la busqueda:
Sentencia ECL para obtener TODOS los conceptos cuyo sitio del procedimiento - directo es igual a la estructura uterina:
# *: 405813007 |sitio del procedimiento - directo (atributo)| = 35039007 |estructura uterina (estructura corporal)|
el texto entre pipes "|" es ignorado por el compilador.

```
import sys
import requests
import json
from requests.structures import CaseInsensitiveDict
url_Final= "https://snowstorm.msal.gov.ar:443/MAIN/concepts?activeFilter=true&termActive=true&ecl=*%3A%20405813007%20%7Csitio%20del%20procedimiento%20-%20directo%20(atributo)%7C%20%3D%2035039007%20%7Cestructura%20uterina%20(estructura%20corporal)%7C&offset=0&limit=10"
url = url_Final
headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"
headers["Accept-Language"] = "es"

resp = requests.get(url, headers=headers)
print ("Estado del Servidor: ")
print (resp.status_code)
print ("Resultado de la búsqueda refinada por ECL: ")
print(resp.text)
diccionario=json.loads(resp.text)
print(diccionario)
print("------------------------------------------------------")
claves= diccionario.keys()
print(claves)
print("------------------------------------------------------")
print(diccionario["items"])
print("------------------------------------------------------")
print(type(diccionario["items"]))
print("------------------------------------------------------")
x = diccionario["items"]
print(x[:])
```
